## 주제 : 날씨에 따라 여행가기 가장 적합한 날이나 지역을 추천해 주는 프로그램 제작

                                                                                                   2019101230 신희연

### 1. 프로젝트 목표 및 내용 :

이 프로젝트의 목표는 이용자들에게 여행을 가기 좋은 날씨를 가진 날짜를 알려주는 역할을 하는 것이다. 이 프로그램은 날짜를 알려주어 비교적 높은 확률로 여행을 갈 때 이용자가 원하는 날씨를 맞이할 수 있을 것이다. 이 프로그램은 최근 10년간 기상청에서 모은 날씨 실측 데이터를 바탕으로 선택한 지역의 원하는 날씨의 여행에 가장 적합한 여러 날짜를 추천해 준다. 날짜는 맞는 범위에 따라 하루가 될 수도, 1년 중 20일이 될 수도 있다. 그 범위에 들어가는 모든 날짜를 사용자에게 구체적인 날씨와 함께 보여주어 사용자가 적합한 날짜를 고르도록 한다.

이 프로그램의 역할은 이용자가 특정 지역을 정하고 날씨 종류마다 원하는 범위를 정하면 여행지에 맞춰서 일 년 동안 가장 사용자가 원하는 날씨 범위 안에 있는 날짜를 추천해 준다. 구체적으로 사용자가 선택할 수 있는 것은, 예를 들어, 습도가 높았으면 좋겠다, 바람이 많이 불었으면 좋겠다. 등이 될 수 있다.

### 2. 주제 선정 이유 또는 이 프로젝트의 필요성 및 주제 변경 이유 :

여행을 갈 때, 더 저렴한 비용과 체계적으로 계획을 세우기 위해 미리 여행을 세우고 예약을 하는 경우가 많다. 여행을 미리 세우다 보면, 몇 달 뒤나 몇 주 뒤에 날씨는 구체적으로 나와 있지 않고 대략적인 정보만 제공해 주고, 가끔은 그 날짜가 다가오면 날씨가 아예 바뀌어 버리는 경우가 있다. 이럴 때, 여행을 다 계획을 해놓았지만, 날씨가 맞지 않아 곤란한 경우가 많이 발생한다. 

주변 사람들의 경우, 그런 경우가 많아서 다양한 사람들의 더욱 성공적인 여행을 위해 이러한 주제로 정하게 되었다. 연구 가설 및 연구 가설의 도출 근거는 과거 10년간의 실적 데이터를 토대로 통계학적 분석을 통한 추천을 하는 것이다. 기상청의 지난 데이터를 토대로 이용자들이 효과적으로 날짜를 선정할 수 있다. 이 프로그램의 연구가설은 이용자가 원하는 지역과 원하는 날씨 특징을 선택하면, 기존에 존재하는 데이터를 종합한다. 

원래의 계획은, 이용자들이 지역뿐만 아니라 날짜 범위를 선택할 수 있었다. 하지만, 기본적인 주제는 같지만 이용자들이 날짜 범위는 선택하지 않고 지역을 선택할 경우 1년 중 원하는 날짜를 알려주는 프로그램만 만들게 되었다. 그 이유는 사용자가 1년 전체 데이터에서 나오는 여러 날을 기반으로 직접 선택할 수 있게 날짜 선택의 폭을 1년으로 확대하기 위해서이다. 날짜 범위를 한정적으로 좁히게 되면, 사용자가 원하는 날씨를 가지고 있는 날짜를 알려주기 어려워진다.


### 3. 데이터 획득 :

기상청에서의 2010년부터 취합된 기상 통계자료에서 대한민국 지역(총 102개 지점)의 일별 날씨를 획득하며 모은다. 기온,강수,강설,기압,바람세기,습도등의 자료를 지역별로 취합한다.
1. 데이터셋 소스 : 기상청에서 제공하는 종관기상관측(ASOS) 데이터셋을 활용
2. 데이터셋 제공요소 : 기온, 강수, 바람, 기압, 습도, 일사, 일조, 눈, 구름, 시정, 지면 상태, 지면 · 초상 온도, 등 
3. 제공 자료 형태 : 분, 시간, 일, 월등의 데이터셋을 제공하며 분,시간은 너무 방대한 관계로 일일 데이터를 전국을 대상으로 최근 10년간 자료로 검색된다. 또한, 모든 시와 도에서 적어도 한 지역의 데이터는 수집한다. (CSV 형식으로 230건)
    4. 하나의 데이터만 있는 서울, 대구, 부산, 인천, 광주, 대전, 울산, 제주는 그 데이터를 사용하였고, 여러 개의 지역으로 나누어져 있는 도 중, 강원도(강릉, 속초, 홍천), 경기도(수원, 이천, 파주), 전라남도(보성군, 여수), 충청북도(청주), 충청남도(부여), 전라북도(전주), 경상북도(울릉도, 포항), 경상남도(거제, 통영)는 여행지로 유명한 지역으로 1-3 지역의 데이터를 선택한다. (제공지점 : 총 23개 지역)


### 4. 구현 내용 설명 :
1. 데이터 획득 : 3번에서 획득한 기상청 포털데이터는 메터데이터형식으로 지역별 매 1년간 CSV 형태의 zip 압축파일이므로 총 23개의 지역의 총 10년씩의 230건을 획득한다.
2. 데이터 가공 및 추출 방법: 
    3. 소프트웨어를 사용하기 전, 230개의 데이터를 모두 zip 압축파일을 풀어 CSV 형식으로 만든 후, 엑셀을 이용해 모두 하나의 데이터로 합친다. (이 파일 이름은 SURFACE_ASOS_DAY_ALL_E2.csv 로, 파일에 함께 첨부되었다.)
        4. 이때, 헤더는 데이터내에서 영어로 바꾼다.
    5. 제공된 내용 중 평균/최저/최고기온, 일 강수량, 평균 풍속, 평균 상대습도, 합계 일사, 일 최심신적설 (총 8 종류)를 추출할 것이며 사용된 소프트웨어는 파이선에 추가 라이브러리를 사용한다.
            *원래 계획은 최저, 최고 기온으로 일교차를 계산하고 합계 일사와 일 최심 신적설을 함께 추출하여 사용하려 했지만, 값이 존재하지 않거나 0으로 처리된 지역이나 연도가 너무 많아 추가하지 않게 되었다. 이 내용들을 포함했다면, 원하는 지역이 단지 데이터가 없어서 0으로 표시되어 제외될 수 있기 때문이다.
        6. 8 종류의 영어 이름은 AVG_temp(°C), day_precipitation(mm), AVG_wind(m/s), AVG_relative_hum(%)  이다.
    7. 파이선을 이용한 전처리 과정을 설명하면,
        8. 데이터 좁히기 : 원하는 8종류를 포함한 새로운 표를 만들어 변수에 저장한다. 
        9. 중복 제거 : 85419건 중, 날짜가 중복되는 1461데이터를 삭제해 총 83958건을 남긴다. 
        10. Null 제거 : 평균 기온, 평균 풍속과 평균 습도에서 총 103건이 지워지고 83855건을 남긴다. 일 강수량은 이 경우에 해당되지 않기에 Null값이 있어도 남겨둔다. 

        11. N/A값을 0으로 바꾸기 : 앞 단계에서 Null 제거를 하지 않고 남긴 값들은 Null 값을 0으로 바꾼다. (여기까지의 과정에서 나온 데이터가 my_first_table.csv로, 파일에 함께 첨부되었다.)
        12. 모든 년도의 평균값 내기 : 10 년도의 평균기온, 평균 풍속, 평균 습도와 일 강수량을 평균값을 낸 뒤, 날짜에 연도는 없애고 달 수과 일 수만 남긴다(201_-10-31 → 10-31). 이때, 평균을 내면서 소수점이 무한으로 나올 수 있는 경우를 생각하여 소수점 2-3번째 자리까지만 남겨둔다. 이 과정을 겪으면, 83855줄의 데이터가 8030줄로 축소된다. (여기까지의 과정에서 나온 데이터가 my_second_table.csv로, 파일에 함께 첨부되었다.)
        13. 새로운 테이블에 범위 나누어 구간 적기 : .median() 함수를 이용해 중간값으로 여러 구간을 만들어 범위를 지정한다. 평균 온도는 1-8 구간이 존재하고, 평균 습도, 평균 풍속과 일 강수량은 1-2 구간이 존재한다. 평균 온도는 계절마다 다르고 범위가 넓기 때문에 사용자가 선택하기 쉽게 여덟 구간으로 나누었고, 계절에 비교적 영향을 미치지 않고 범위가 넓지 않은 평균 습도, 평균 풍속과 일 강수량은 두 구간으로 나누었다. (여기까지의 과정에서 나온 데이터가 my_third_table.csv로, 파일에 함께 첨부되었다.)


3. 사용 소프트웨어 기능 : 
    14. 다량의 데이터를 처리해야 하므로 Pandas를 파이썬에 import 해서 사용하여 기상 데이터를 DataFrame으로 제공자료를 읽어서 추출한다.
    15. 데이터의 빠른 가공 및 통계적 해석을 위해서 median 함수로 구간을 나누어 범위를 지정해 주었다. 
    16. 앞에서 불러온 데이터를 기반으로 한 가지 기능을 만든다.
        17. 그 기능은 사용자가 하나 또는 원하는 지역과 날씨 특징에 따라 원하는 범위를 선택하면, 프로그램이 그 날짜 범위 내에서 가장 적합한 특정한 날을 알려주는 것이다. 
4. 소프트웨어 동작 방법 :
    21. 사용자가 처음으로 프로그램을 돌릴 때, 프로그램에 대한 간략한 설명을 넣는다. (introduction 함수)
    22. 간략한 설명과 함께 데이터 원본을 3 단계를 통해 가공된다. 이때, 비교적 시간이 오래 걸려 사용자가 어느 부분까지 가공이 되었는지 알 수 있게, 데이터를 가공하는 함수 3개( read_csv 함수, second_step_table 함수, third_step_table 함수)에 출력문을 넣어 어느 과정까지 끝났는지 알 수 있게 한다. (예시. reading original csv…     done for the first-table)
    23. 가공이 끝나면, 이 프로그램의 첫 번째 기능은, 사용자는 먼저 총 23개의 지역 중에 사용자가 원하는 지역을 선택하는 것이다. 지역과 지역에 해당하는 숫자가 함께 주어지는데, 지역에 맞는 숫자를 입력값으로 넣어야 한다. 그리고 이 입력값을 반환한다. 이때, 주어진 지역이 아닌 다른 값을 입력했을 때에는 알맞은 지역을 선택할 때까지 사용자에게 계속 묻도록 설정되었다. (get_region 함수)
    24. 그다음, 총 4가지의 날씨 특징을 입력한다. 이때, 날씨 특징은 평균 온도는 1부터 8까지 중에서, 평균 습도, 평균 풍량과 일 강수량은 1부터 2까지 (낮음, 높음)으로 주어지면 사용자가 이 범위의 숫자 중 고르는 것이다. 소수점으로 고를 수 없으며 범위 안에 있는 정수를 적어준다. 이때, 온도 같은 경우에는 각 구간마다 해당되는 범위에 속하는 온도를 알려주어(예시. section 2 is between 1.48 (°C) and 5.105 (°C) ), 사용자가 보다 쉽게 선택할 수 있게 도와준다. 이러한 구간 값들을 입력받아 반환하여 리스트로 정리해두어 추후 사용한다. 이때, 주어진 범위 밖의 값을 입력했을 때에는 알맞은 범위 값을 입력할 때까지 사용자에게 계속 묻도록 설정되었다. (get_input 함수)
    25. 이렇게 모든 데이터를 입력받으면, get_region와 get_input으로부터 입력받은 지역과 구간에 맞는 날짜만 있는 row를 합쳐서 my_third_table을 기반하여 새로운 테이블을 만든다. 이때, 입력받은 지역과 평균 온도는 우선순위로 들어가 새로운 테이블에 무조건 반영된다. 그다음으로 반영되는 순위는 일 강수량, 평균 풍속, 평균 습도 순이다. 일 강수량, 평균 풍속, 평균 습도는 사용자가 원하는 구간의 데이터가 우선순위에 기반한 데이터에 존재하는지 확인하고, 존재한다면 그에 따라 데이터를 좁힌다. 만약 일 강수량, 평균 풍속, 평균 습도 중 선택한 구간의 값이 존재하지 않는다면 그전에 만들어진 테이블을 반환한다. (get_user_table 함수)
    26. 마지막으로, get_user_table 함수에 기반하여 만들어진 최종 테이블에 포함된 날짜, 평균 온도, 일 강수량, 평균 풍량, 평균 습도를 출력해준다. 결과적으로, 프로그램이 사용자가 정한 범위에 따른 가장 적합한 특정한 날을 알려준다. (result_for_user 함수)
    27. 예시로는, 사용자가 원하는 지역으로 울산을 선택하고, 평균 온도 구간은 3(between 5.105(°C) and 9.95(°C))을 선택하고, 일 강수량은 1, 평균 풍속은 1, 마지막으로 평균 습도는 1을 고른다면, 적합한 날짜로 3월 17일을 출력한다. 정확히 출력되는 문장은 “the result date is 03-17, average temperature is 8.73(°C), day precipitation is 0.85(mm), average wind is 2.11(m/s), and average relative humidity is 45.34(%)” 이다.




### 5. 구현 내용 설명 :
    1. 원본 데이터를 분석하여 값이 들어가지 않은 데이터는 제외하고, 중간값을 낼 수 있고 예외적이지 않는 데이터를 만들었다. 평균값으로 총합을 개수로 나누는 ‘Average’는 사용하지 않고 ‘Median’인 중간값을 사용하여 예외적인 숫자들은 배제하는 방식으로 데이터를 좁혀갔다. 총 3가지의 가공된 데이터, csv 파일을 결과로 첨부하였다. 마지막으로  가공된 데이터에서는 중간값을 4번 이용하여 총 8구간을 만든다. 구간을 또한, 마지막으로 사용자가 선택한 구간과 지역에 따라 마지막 데이터가 만들어진다. 
    2. 또한, 사용자가 원하는 지역의 1년 전체의 데이터를 보여주는 그래프를 결과로 보여준다. 사용자가 선택한 구간의 날짜와 그에 따른 날씨를 알려주는 것 또한 중요하지만, 1년의 그 지역의 전체적인 추세를 보여주어 사용자가 그 지역에 대한 날씨 정보를 그래프를 통해서 보여준다. 소스 코드에 있는 예시는 강릉이다.


### 6. 결론:
1. 지역에 따라 원하는 날씨를 선택하면 그 날씨에 맞는 날씨를 추천해 주는 프로그램을 만들 수 있었다. 10년 데이터를 기반으로 사용자들이 보다 성공적인 여행을 할 수 있게 돕는 것이 이 프로그램의 목표였기에, 이 프로젝트의 목표를 이루었다고 생각한다. 10년 데이터의 값을 평균을 내고 예외적인 값은 포함하지 않아 정확도를 높였다. 본 프로그램을 통해 과거 데이터를 바탕으로 통계학적인 예측을 통해 실제 날씨를 예측을 할 수 있고, 보다 구체적인 날짜를 알려줄 수 있다는 것을 확인하였다. 
2. 비록, 모든 날씨 특징들, 예를 들어서 합계 일사량, 일교차 등을 포함하여 사용자가 원하는 정확한 날씨를 반영하는 것에는 한계가 있었다. 생각보다 지역마다 존재하지 않는 데이터가 많았고, 날짜가 중복되는 경우도 있었다. 하지만, 이러한 예상하지 못했던 데이터 결과에도 불구하고 최대한 반영할 수 있는 날씨 특징을 모아 4가지 날씨 특징 내에서 사용자가 여행하기 좋은 날짜를 알려줄 수 있었다. 또한, 다량의 데이터를 소화하기 때문에 프로그램이 예상보다 작동하는데 많은 시간이 소요되었다. 그래서, 한 번 데이터가 가공되면, 그 이후로 사용자가 입력할 때마다 새로 가공되지 않도록 코드를 나누었다. 이와 같이 처음 계획서를 작성할 때 예상했던 결과와는 조금 다르게 나왔지만, 전체적은 틀은 바뀌지 않았다.
3. 또한, 1년 전체를 범위로 두어, 단지 한정된 날짜 범위 이내가 아니라 비슷한 날씨를 가진 다양한 날짜 중에서 직접 고를 수 있다는 편리성 또한 가졌다. 실제로, 사용자에게 여행하기 좋은 날짜를 추천해 주는 프로그램이라서, 한정된 범위 내에서 사용자에게 원하는 날짜가 있다고 하더라도 그 날짜에 여행을 가지 못하는 경우도 있다. 이러한 경우를 위해 범위를 넓게 두어 보다 효과적으로 여행을 갈 수 있도록 도왔다. 
4. 마지막으로, 이 프로그램을 기반으로 더 많은 지역을 포함하고, 10년 간의 데이터가 아닌 수십 년간, 또한 지금부터의 날씨를 축적하여 날씨를 예측하면 보다 정확하게 원하는 날짜를 알려줄 수 있을 것이다.
